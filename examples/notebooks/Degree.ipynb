```python
import networkx as nx
""" I am using a modified networkx version, to support weighted global efficiency"""
import load_connectivity 
import copy

import matplotlib
import graphing

matplotlib.rcParams['figure.figsize'] = [15,5]

print nx.__version__

network = load_connectivity.load_2015_connectivity_matrix()
```

# Degree 

```python
degree = list(network.degree())
degree_dict = {}

for d in degree:
    degree_dict[d[0]] = d[1]
graphing.plot_measurement(degree_dict ,degree_dict.keys(),"Degree")
```

# Closeness

```python
closeness = nx.algorithms.closeness_centrality(network, distance='weight')

graphing.plot_measurement(closeness, "Closeness")
```
